###Práctica 1

Indicadores del Banco Mundial

```
#Importamos las librerías, aquí necesitamos la más importante que es request y json
import requests
import json
import pandas as pd

# Obtenemos todas las fuentes de la API del Banco Mundial
sources = requests.get("http://api.worldbank.org/v2/sources?per_page=100&format=json")
sourcesJSON = sources.json()
sourcesJSON
```

```
# Analizamos la respuesta para ver los nombres de las fuentes y los id's
for i in sourcesJSON[1]:
    if i["name"] == "International Debt Statistics":
        print("The source ID for International Debt Statistics is " + i["id"])
    else:
        pass
    print(i["id"],i["name"])
```

```
# Solicitamos los indicadores para el tema Deuda Externa
indicators = requests.get("http://api.worldbank.org/v2/indicator?format=json&source=6")
indicatorsJSON = indicators.json()
indicatorsJSON
```

```
# Imprimimos el número total de indicadores en este tema
print("Hay" + str(indicatorsJSON[0]["total"]) + "indicadore IDS")
```

```
# Obtenga todos los indicadores de deuda externa, con un parámetro por página de 500.
indicators = requests.get("http://api.worldbank.org/v2/indicator?format=json&source=6&per_page=500")
indicatorsJSON = indicators.json()
indicatorsJSON
```

```
# Analizamos la respuesta para ver los identificadores (ID's) y nombres de los indicadores
for i in indicatorsJSON[1]:
    IDSindicators = (i["id"],i["name"])
    print(IDSindicators)
```

```
#Buscamos  "DT.DOD.DLXF.CD External debt stocks, long-term (DOD, current US$)."
# Utilizamos el código del indicador para definir la variable "indicador"
indicator = "DT.DOD.DLXF.CD"

# Analice la respuesta para obtener la "sourceNote" o la definición del indicador deseado
for dict_entity in indicatorsJSON[1]:
    if dict_entity["id"] == indicator:
        print(dict_entity["sourceNote"])
    else:
        pass
```

###Obtenemos información de un país

```
# Solicitamos (requests) las ubicaciones
dlocations = requests.get("http://api.worldbank.org/v2/sources/6/country?per_page=300&format=JSON")
dlocationsJSON = dlocations.json()

# Analicemos la respuesta para ver los ID y nombres de ubicación
dlocations = dlocationsJSON["source"][0]["concept"][0]["variable"]
listLen = int(len(dlocations))

# Creamos un df con los valores de los lugares
df = pd.DataFrame(columns=["id", "value"])     
for i in range(0,listLen):
    code = dlocations[i]["id"]
    name = dlocations[i]["value"]
    df = df.append({"id":code, "value":name}, ignore_index = True)
dlocationsList = df

# Vemos los primeros renglones del dataframe
print(dlocationsList)
```

```
#Aquí solo obtenemos todos los lugares y conjuntos (continentes, etc)
# Solicitamos (requests) las ubicaciones
clocations = requests.get("http://api.worldbank.org/v2/sources/6/counterpart-area?per_page=300&format=JSON")
clocationsJSON = clocations.json()

# Analicemos la respuesta para ver los ID y nombres de ubicación
clocations = clocationsJSON["source"][0]["concept"][0]["variable"]
listLen = int(len(clocations))

# Creamos un df con los valores de los lugares
df = pd.DataFrame(columns=["id", "value"])     
for i in range(0,listLen):
    code = clocations[i]["id"]
    name = clocations[i]["value"]
    df = df.append({"id":code, "value":name}, ignore_index = True)
clocationsList = df

# Vemos los primeros renglones del dataframe
print(clocationsList)
```

```
# Realizamos la petición a la API e instanciamos la clase DataFrame y lo designamos como "EXD"
EXD = wbdata.get_dataframe(indicatorSelection, country = locationSelection, data_date = timeSelection, convert_date=True)
EXD
```

###Modificar la presentación

```
#Creamos otro DF pero solo con el indicador de deuda que seleccionamos anteriormente
EXD_only_debt = wbdata.get_dataframe(indicator2, country = locationSelection, data_date = timeSelection, convert_date=True)
EXD_only_debt
```

```
# Cambiamos la forma del DF
EXDreshaped = pd.DataFrame(EXD_only_debt.to_records())

# Creamos una función que cambiará las unidades a miles de millones y aplicaremos redondeo
def formatNum(x):
    y = x/1000000000
    z = round(y)
    return(z)

# Aplicamos la función en la columna que queramos, en este caso, la única
EXDreshaped.ExternalDebtStock = formatNum(EXDreshaped.ExternalDebtStock)
```

```
# Renombramos las columnas
EXDclean = EXDreshaped.rename(index=str, columns={
    "date":"Year",
    "country":"Region",
})
```

```
# Removemos la cita "(excluding high income)" por cada nombre de las regiones
EXDclean["Region"] = EXDclean["Region"].str.replace("excluding high income","").str.replace(")","").str.replace("(","")
```

```
#Aplicamos metodo head para mostrar las primeras rows
print(EXDclean.head())
```

###Gráfica sencilla con código facilito
